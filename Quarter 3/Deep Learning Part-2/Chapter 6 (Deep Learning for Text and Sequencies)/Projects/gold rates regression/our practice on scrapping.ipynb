```python
class Cordinate(object):
    
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def square(self):
        return self.x**2,self.y**2
    
    def distance(self,other):
        
        x_diff_sq = (self.x - other.x)**2
        y_diff_sq = (self.y - other.y)**2
        
        return (x_diff_sq+y_diff_sq) ** 0.5
```

```python
a = Cordinate(7,8)
b = Cordinate(6,5)
```

```python
from bs4 import BeautifulSoup as bs
import requests

print()

covid_link = "https://www.worldometers.info/coronavirus/#countries"

req = requests.get(covid_link)

req = req.text
soop = bs(req,"html.parser")

table = soop.find("table",{"id":"main_table_countries_today"})

name = input("Enter Country name: ")

for i in table.find_all("tr"):
    for j in i.find_all("a"):
        for k in j:
            if k == name.capitalize():
                m = k
                print()
                print(" Extracting......")
                print()

                parent = (m.parent.parent.parent).getText()
                parent_li = parent.split("\n")

                parent_li.pop(0)
                parent_li.pop(-1)

                thead = soop.find("thead")
                thead_li = thead.getText().split("\n")

                thead_li.pop(0)
                thead_li.pop(0)
                thead_li.pop(-1)
                thead_li.pop(-1)

                thead_li[15] = "Continent"
                thead_li[1] ="Country"

                for i in range(len(parent_li)):
                    print("",thead_li[i],":",parent_li[i])
else:
    print("Country Not Found")
```
