```python
def scrap1(link):
    
    source = requests.get(link, headers=headers).text
    soup = bs(source,"html.parser")
    
    for i in soup.find_all("tr"):
        
        date = i.find("td",attrs={'width':'90'})
        rate = i.find("td",attrs={'width':'60'})
        
        if date == "" or rate == "" :
            
            date = i.find("td",attrs={'width':'84'})
            rate = i.find("td",attrs={'width':'75'})

        if date == None or rate == None:
            continue
        else:
            dates.append(date.getText())
            rates.append( float(rate.getText()) )
    #         print(f"{date.getText()} : {rate.getText()}")
    
    return dates,rates

def scrap2(link):
    
    source = requests.get(link, headers=headers).text
    soup = bs(source,"html.parser")
    
    for i in soup.find_all("tr"):
        date = i.find("td",attrs={'width':'84'})
        rate = i.find("td",attrs={'width':'75'})

        if date == None and rate == None:
            continue
        else:
            dates.append(date.getText())
            rates.append( float(rate.getText()) )
    #         print(f"{date.getText()} : {rate.getText()}")
    
    return dates,rates
```

```python
import re
dates, rates = [], []
def extractData(l):
    result1, result2 = re.findall("\d+\.\d+",l) , re.sub("\d*\.\d+","",l)
    rates.append(result1)
    dates.append(result2)
    return dates, rates
```

```python
for i in data:
    for j in i:
        dates,rates = extractData(j)
```
